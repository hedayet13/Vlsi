#N-mos Transistor i-v Characteristics Curve


```python
'''
N-mos transistor I-V characteristic

t_ox= oxidation thickness
mu= mobility constant
Vt=threshold voltage
Vgs=gate to source voltage. Range of (0 to 5)
Vds=drain to source voltage
Ids = Drain to source current
W= Gate width
L= Gate Length 
'''
## importing all essential library
import matplotlib.pyplot as plt
import numpy as np
 
# all Variables
t_ox = 100*(10**-8)   
mu=350
Vt=0.7 
range_Vgs = range(1,6)
range_Vds= np.linspace(0,5)
W=4
L=2
epsilon = 8.854*(10**-14)
 
# Beta value Execution
Beta = mu*(3.9*epsilon/t_ox)*(W/L)
 
# Showing the curve
plt.figure(figsize=(10,7))
for Vgs in range_Vgs:
  Ids_value = []
  for Vds in range_Vds:
    if Vgs<Vt:
      Ids=0
    elif Vgs>Vt and Vds<=(Vgs-Vt):
      Ids= Beta*(Vgs-Vt-(Vds/2))*Vds
    elif Vgs>Vt and Vds>(Vgs-Vt):
      Ids=(Beta/2)*((Vgs-Vt)**2)
    Ids_value.append(Ids*1000)
  plt.plot(range_Vds,Ids_value,label = "Vgs={}".format(Vgs))
plt.ylim([0, 2.5])
plt.title(r'n-mos transistor I-V characteristics curve')
plt.xlabel('Vds (Voltage)')
plt.ylabel('Ids (mA)')
plt.legend()
plt.show()
```

#P-mos transistor I-v Characteristics curve

```python
'''
P-mos transistor I-V characteristic

'''
## importing all essential library
import matplotlib.pyplot as plt
import numpy as np

# all Variable 
t_ox = 100*(10**-8)
mu=120
Vt=-0.7
range_Vgs = range(-5,0)
range_Vds= -np.linspace(0,5)
W=4
L=2
epsilon = 8.854*(10**-14)

# Beta value Execution
Beta = mu*(3.9*epsilon/t_ox)*(W/L)

# Showing the curve
plt.figure(figsize=(10,7))
plt.tick_params(axis="y",which='both',labelleft=False,labelright=True)

for Vgs in range_Vgs:
  Ids_value = []
  for Vds in range_Vds:
    if Vgs>Vt:
      Ids=0
    elif Vgs<Vt and Vds>=(Vgs-Vt):
      Ids= -Beta*(Vgs-Vt-(Vds/2))*Vds
    elif Vgs<Vt and Vds<(Vgs-Vt):
      Ids=-(Beta/2)*((Vgs-Vt)**2)
    Ids_value.append(Ids*1000)
  plt.plot(range_Vds,Ids_value,label = "Vgs={}".format(Vgs))
plt.title(r'p-mos transistor I-V characteristics curve')
plt.xlabel('Vds(Voltage)')
plt.ylabel('Ids (mA)',labelpad=-625)
plt.legend()
plt.show()
```

```python
'''
t_ox= oxidation thickness
mu= mobility constant

Vtn=threshold voltage of n-mos
Vtp=threshold voltage of p-mos

Vgsn=gate to source voltage(n-mos) 
Vgsp=gate to source voltage(p-mos)

Vdsn=drain to source voltage(n-mos)
Vdsp=drain to source voltage(p-mos)

Idsn = Drain to source current(n-mos)
Idsp = Drain to source current(p-mos)

W= Gate width
L= Gate Length 
'''
## importing all essential library
import matplotlib.pyplot as plt
import numpy as np
 
# all Variables
t_ox = 100*(10**-8)   
mu=350

Vtn=0.7 
Vtp=-0.7
range_Vin = range_Vgsn = range(1,6)
range_Vgsp = range(-5,0)
range_Vout = range_Vdsn= np.linspace(0,5)
range_Vdsp = -np.linspace(0,5)
W=4
L=2
epsilon = 8.854*(10**-14)
 
# Beta value Execution
Beta = mu*(3.9*epsilon/t_ox)*(W/L)
 
# Showing the curve
plt.figure(figsize=(10,7))
for Vgsn in range_Vgsn:
  Idsn_value = []
  for Vdsn in range_Vdsn:
    if Vgsn<Vtn:
      Idsn=0
    elif Vgsn>Vtn and Vdsn<=(Vgsn-Vtn):
      Idsn= Beta*(Vgsn-Vtn-(Vdsn/2))*Vdsn
    elif Vgsn>Vtn and Vdsn>(Vgsn-Vtn):
      Idsn=(Beta/2)*((Vgsn-Vtn)**2)
    Idsn_value.append(Idsn*1000)
  plt.plot(range_Vdsn,Idsn_value,label = "Vgsn={}".format(Vgsn))

for Vgsp in range_Vgsp:
  Idsp_value = []
  for Vdsp in range_Vdsp:
    if Vgsp>Vtp:
      Idsp=0
    elif Vgsp<Vtp and Vdsp>=(Vgsp-Vtp):
      Idsp= -Beta*(Vgsp-Vtp-(Vdsp/2))*Vdsp
    elif Vgsp<Vtp and Vdsp<(Vgsp-Vtp):
      Idsp=-(Beta/2)*((Vgsp-Vtp)**2)
    Idsp_value.append(Idsp*1000)
  plt.plot(range_Vdsp,Idsp_value,label = "Vgsp={}".format(Vgsp))
# plt.ylim([-2.5, 2.5])
# ply.yticks(-2.5,2.5,.5)
# start,end = plt.ylim()
plt.yticks(np.arange(-2.5,3,0.5))
plt.xticks(np.arange(-5,6,1))
plt.title(r'Graphical derivation of CMOS inverter DC characteristics')
plt.xlabel('Vds (Voltage)')
plt.ylabel('Ids(mA)')
plt.axhline(y=0,color='dimgray')
plt.axvline(x=0,color='dimgray')
plt.grid()
plt.legend()
plt.show()
```
