#N-mos Transistor i-v Characteristics Curve


```python
'''
N-mos transistor I-V characteristic
t_ox= oxidation thickness
mu=
Vt=threshold voltage
Vgs=gate to source voltage. Range of (0 to 5)
Vds=drain to source voltage
Ids = Drain to source current
W= Gate width
L= Gate Length
epsilon = 
'''
## importing all essential library
import matplotlib.pyplot as plt
import numpy as np

# all Variables
t_ox = 100*(10**-8)   
mu=350
Vt=0.7 
range_Vgs = range(1,6)
range_Vds= np.linspace(0,5)
W=4
L=2
epsilon = 8.854*(10**-14)

# Beta value Execution
Beta = mu*(3.9*epsilon/t_ox)*(W/L)

# Showing the curve
plt.figure(figsize=(10,7))
for Vgs in range_Vgs:
  Ids_value = []
  for Vds in range_Vds:
    if Vgs<Vt:
      Ids=0
    elif Vgs>Vt and Vds<=(Vgs-Vt):
      Ids= Beta*(Vgs-Vt-(Vds/2))*Vds
    elif Vgs>Vt and Vds>(Vgs-Vt):
      Ids=(Beta/2)*((Vgs-Vt)**2)
    Ids_value.append(Ids*1000)
  # print(Ids_value,range_Vds)

  plt.plot(range_Vds,Ids_value,label = "Vgs={}".format(Vgs))
plt.ylim([0, 2.5])
# ymin, ymax = plt.ylim()
# plt.ylim(ymin * scale_factor, ymax * scale_factor)
plt.title(r'N-mos Transistor I-V characteristics')
plt.xlabel('Vds (Volage)')
plt.ylabel('Ids (mA)')
plt.legend()
plt.show()
```

#P-mos transistor I-v Characteristics curve

```python
'''
P-mos transistor I-V characteristic

'''
## importing all essential library
import matplotlib.pyplot as plt
import numpy as np

# all Variable 
t_ox = 100*(10**-8)
mu=350
Vt=-0.7
range_Vgs = range(-5,0)
range_Vds= -np.linspace(0,5)
W=4
L=2
epsilon = 8.854*(10**-14)

# Beta value Execution
Beta = mu*(3.9*epsilon/t_ox)*(W/L)

# Showing the curve
plt.figure(figsize=(10,7))
for Vgs in range_Vgs:
  Ids_value = []
  for Vds in range_Vds:
    if Vgs>Vt:
      Ids=0
    elif Vgs<Vt and Vds>=(Vgs-Vt):
      Ids= -Beta*(Vgs-Vt-(Vds/2))*Vds
    elif Vgs<Vt and Vds<(Vgs-Vt):
      Ids=-(Beta/2)*((Vgs-Vt)**2)
    Ids_value.append(Ids)
  

  plt.plot(range_Vds,Ids_value,label = "Vgs={}".format(Vgs))

plt.title(r'N-mos Transistor I-V characteristics')
plt.xlabel('Vds(Volage)')
plt.ylabel('Ids(Ampere)')
plt.legend()
plt.show()
```
